```python
import pandas as pd

df = pd.read_csv('housing.csv')
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = ['median_house_value', 'median_income']
sns.pairplot(df[cols])
plt.show()
```

```python
sns.heatmap(df[cols].corr(), annot=True)
plt.show()
```

```python
from re import S
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

x= df['median_income'].values.reshape(-1,1)
y= df['median_house_value'].values.reshape(-1,1)

ss_x=StandardScaler()
ss_y=StandardScaler()

x_ss = ss_x.fit_transform(x)
y_ss = ss_y.fit_transform(y)

lr = LinearRegression()
lr.fit(x_ss,y_ss)
```

```python
import numpy as np

ingreso_medio = float(input("Ingrese un valor de ingreso anual medio: "))

ingreso_medio_std = ss_x.transform([[ingreso_medio]])

result_std = lr.predict(ingreso_medio_std)

result = ss_y.inverse_transform(result_std)

print("Prediccion (en miles de dolares):", round(result[0][0]), "USD")
```
